```python
!pip install faker
import pandas as pd
import random
from faker import Faker
from datetime import datetime, timedelta

# Create a Faker instance
fake = Faker()

# Generate 10,000 rows of fake data
data = []

# To keep track of used serial numbers
used_serial_nos = set()

# Start the serial number from 1
serial_no = 0

while len(data) < 10000:
    # Increment the serial number
    serial_no += 1
    
    # Generate a unique serial_no
    while serial_no in used_serial_nos:
        serial_no += 1
    used_serial_nos.add(serial_no)

    # Date Created
    date_created = fake.date_between(start_date="-6m", end_date="today")
    
    # Service
    service = fake.random_element(elements=("Service A", "Service B", "Service C"))
    
    # Client Name
    client_name = fake.company()
    
    # Client Group
    client_group = fake.random_element(elements=("Group 1", "Group 2", "Group 3"))
    
    # Location Sector
    location_sector = fake.random_element(elements=("Private Equity", "Financial Services", "Technology", "Healthcare"))
    
    # Country
    country = fake.country()
    
    # Opportunity Status
    opportunity_status = fake.random_element(elements=("Win", "Proposal", "Opportunity", "Preliminary", "Lost", "Abandoned"))
    
    # Opportunity Size (INR)
    opportunity_size_inr = fake.random_int(min=100000, max=10000000)
    
    # Opportunity Size (INR lacs)
    opportunity_size_inr_lacs = opportunity_size_inr / 100000
    
    # Opportunity Value Range
    opportunity_value_range = fake.random_element(elements=("0-1M", "1M-5M", "5M-10M", "10M-50M", "50M+"))
    
    # Winning Probability
    winning_probability = fake.random_element(elements=("Low", "Medium", "High"))
    
    # Opportunity Services
    opportunity_services = fake.random_elements(elements=("Service A", "Service B", "Service C"), unique=True, length=random.randint(1, 3))
    
    # Services Offered
    services_offered = fake.random_elements(elements=("Service X", "Service Y", "Service Z"), unique=True, length=random.randint(1, 3))
    
    # Client Contact
    client_contact = fake.name()
    
    # Designation Email
    designation_email = fake.email()
    
    # Introduced By
    introduced_by = fake.random_element(elements=(client_contact, "Referral", "Marketing"))

    # Additional Columns
    # Revenue (Assuming revenue ranges from 100,000 to 10,000,000 INR)
    revenue = fake.random_int(min=100000, max=10000000)
    
    # Engagement Partner
    engagement_partner = fake.name()
    
    # Director
    director = fake.name()

    data.append([
        serial_no, date_created, service, client_name, client_group, location_sector, country, opportunity_status,
        opportunity_size_inr, opportunity_size_inr_lacs, opportunity_value_range, winning_probability,
        opportunity_services, services_offered, client_contact, designation_email, introduced_by,
        revenue, engagement_partner, director
    ])

# Create a DataFrame from the generated data
columns = [
    "Serial No", "Date Created", "Service", "Client Name", "Client Group", "Location Sector", "Country",
    "Opportunity Status", "Opportunity Size (INR)", "Opportunity Size (INR lacs)", "Opportunity Value Range",
    "Winning Probability", "Opportunity Services", "Services Offered", "Client Contact", "Designation Email",
    "Introduced By", "Revenue", "Engagement Partner", "Director"
]

df = pd.DataFrame(data, columns=columns)
df.head()

# Save the DataFrame to a CSV file
df.to_csv("fake_dataset.csv", index=False)
```
